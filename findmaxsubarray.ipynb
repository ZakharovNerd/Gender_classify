```python
def findMaxSubArray(A):
    run=A[0]
    res=A[0]
    start=0
    end=0
    if len(A)>1:
        for i in range(1,len(A)):
            if run+A[i]>A[i]:
                run=run+A[i]
            else:
                run=A[i]
                start=i
            if res<run:
                res=run
                end=i
    return A[start:end+1]

findMaxSubArray([-2,1,-3,4,-1,2,1,-5,4])
```
